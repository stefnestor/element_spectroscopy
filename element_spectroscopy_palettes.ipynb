```
for element_name in element:
  e = element[element_name]
  l = len(e)
  for n in range(l):
    s = e[n]
    plt.subplot(l,1,n+1)
    plt.hist(s[0],1,facecolor=s[1],alpha=.9)
    plt.axis("off")
  plt.title(element_name,rotation='vertical',x=-0.1,y=0)
  plt.axis("off")
  plt.show()
  print("\n\n")

```
